### Total number of fundings done in each year.

```python
import pandas as pd
import matplotlib.pyplot as plt
obj=pd.read_csv('startup_funding.csv')
ax=obj['Date']
dx=bx=cx=0
for i in range(0,len(ax)):
    str=ax.iloc[i]
    if str[-2:]=='17':
        dx=dx+1
    elif str[-2:]=='16':
        bx=bx+1
    else:
        cx=cx+1
a=[dx,bx,cx]
b=[2017,2016,2015]
plt.plot(b,a)
plt.scatter(b,a)
plt.show()
print(2015,cx)
print(2016,bx)
print(2017,dx)
```
